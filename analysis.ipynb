## vehicle access stuff

i thought having a car would mean you can get to a grocery store but apparently not??

```python
# millions of people
has_car = 17.4
no_car = 2.1

print(f"people with cars far from stores: {has_car} million")
print(f"people without cars far from stores: {no_car} million")
```

```python
# what percent have cars
total = has_car + no_car
has_car / total * 100
```

so like 89% of people in food deserts actually have cars. wild
